```python
from ucimlrepo import fetch_ucirepo 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import seaborn as sns
import scipy
from scipy.stats import pearsonr
import sklearn
from sklearn import datasets, linear_model
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

# fetch dataset 
student_performance = fetch_ucirepo(id=320) 

X = student_performance.data.features 
y = student_performance.data.targets 

# variable information 
# print(student_performance.variables) 
```

```python
# Filter out rows in X that contain missing values (if any) in selected columns
data_mean = X.iloc[:, [2,6,7,12,13,14,23,24,25,26,27,28,29]].dropna()

# Use the same index to select rows in y
y_filtered = y.loc[data_mean.index].dropna()

# Now transform and scale data_mean (without losing rows in the selection)
scaler = StandardScaler()
Xs = scaler.fit_transform(data_mean)
```

```python
print("Shape of filtered Xs:", Xs.shape)
print("Shape of filtered y:", y_filtered.shape)
```

```python
# print("Shape of data_mean (filtered X):", data_mean.shape)
# print("Shape of y_filtered (filtered y):", y_filtered.shape)

# scaler = StandardScaler()
# Xs = scaler.fit_transform(data_mean)

# Confirm shape consistency between Xs and y_filtered
# print("Shape of standardized Xs:", Xs.shape)
# print("Shape of y_filtered:", y_filtered.shape)

# print(y_filtered.value_counts())


y_single_target = y_filtered.iloc[:, -1]

print("Shape of y_single_target:", y_single_target.shape)


Xs_train, Xs_test, y_train, y_test = train_test_split(
    Xs, y_single_target.values.ravel(), test_size=0.2, random_state=1
)


# Xs_train, Xs_test, y_train, y_test = train_test_split(Xs, y_filtered.values.ravel(), test_size=0.2, random_state=1, stratify=y_filtered)


# Dimensionality Reduction using PCA to reduce features to 6 components
pca = PCA(n_components=6)
pca.fit(Xs_train)

Xs_train = pca.transform(Xs_train)
Xs_test = pca.transform(Xs_test)
```

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
```

```python
svr_model = SVR(kernel='poly')  # You can try 'linear', 'poly', or 'sigmoid' as well

# Fit the model
svr_model.fit(Xs_train, y_train)
```

```python
# Make predictions on the test set
y_pred = svr_model.predict(Xs_test)
regression_score = svr_model.score(Xs_test, y_test)
print("SVM Regression Score:", regression_score)
```

```python
# Calculate Mean Squared Error and R-squared
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
